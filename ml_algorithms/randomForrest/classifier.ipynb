 To understand how random forest works, it is important to understand the techniques that the algorithm is comprised of. I explain them below.

<h3>Bagging</h3>
<p class="selectionShareable">Bagging is a method of generating new datasets from existing data by creating samples of the existing data with replacement. This means there could be repeated values in each of the newly created datasets.</p>
<p class="selectionShareable">Bagging is the magic that makes random forest popular because it avoids overfitting, despite increasing the number of trees. This is because it averages many low-bias and high-variance predictors, thereby reducing the variance without increasing bias. Consequently, random forests can achieve high accuracy without the risk of overfitting or underfitting data. Also, since multiple versions of the dataset are generated, it is possible to work with relatively small datasets.</p>

<h3>Decision Tree Learning</h3>
<p class="selectionShareable">A decision tree learning algorithm can be used for classification or regression problems to help predict an outcome based on input variables. Decision trees are made of:</p>
<ul> 
  <li><p class="selectionShareable"><span class="hs_cos_wrapper hs_cos_wrapper_meta_field hs_cos_wrapper_type_rich_text" data-hs-cos-general-type="meta_field" data-hs-cos-type="rich_text" id="hs_cos_wrapper_post_body" style=""><strong>A root:</strong> The feature that best describes the dataset. This attribute is selected by calculating the Gini index or Information Gain of all the features. There is only one root.</span></p></li> 
  <li><p class="selectionShareable"><span class="hs_cos_wrapper hs_cos_wrapper_meta_field hs_cos_wrapper_type_rich_text" data-hs-cos-general-type="meta_field" data-hs-cos-type="rich_text" id="hs_cos_wrapper_post_body" style=""><strong>Nodes:</strong> Splitting points for decisions.</span></p></li> 
  <li><p class="selectionShareable"><span class="hs_cos_wrapper hs_cos_wrapper_meta_field hs_cos_wrapper_type_rich_text" data-hs-cos-general-type="meta_field" data-hs-cos-type="rich_text" id="hs_cos_wrapper_post_body" style=""><strong>Branches:</strong> Split description.</span></p></li> 
  <li><p class="selectionShareable"><span class="hs_cos_wrapper hs_cos_wrapper_meta_field hs_cos_wrapper_type_rich_text" data-hs-cos-general-type="meta_field" data-hs-cos-type="rich_text" id="hs_cos_wrapper_post_body" style=""><strong>Leaves:</strong> Final-level nodes that cannot be further split.</span></p></li> 
 </ul>

<p class="selectionShareable">Random forest is an ensemble decision tree algorithm because the final prediction, in the case of a regression problem, is an average of the predictions of each individual decision tree; in classification, it's the average of the most frequent prediction. So, the algorithm takes the average of many decision trees to arrive at a final prediction, as shown in the image above.</p>

<h3>When Should You Use Random Forest?</h3>

Furthermore, it is a relatively easy model to build and doesn’t require much hyperparameter tuning. This is because the main hyperparameters are the number of trees in the forest and the number of features to split at each leaf node.</p>

<h2> Random Forest Classifier </h2>

<h2 class="selectionShareable">Step 1: Load Python packages</h2>

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.model_selection import train_test_split
from scipy.io import arff
import pandas as pd
```

<h2 class="selectionShareable">Step 2: Pre-Process the data</h2>
<p class="selectionShareable">For this tutorial, we will be using the ‘<a href="https://archive.ics.uci.edu/ml/datasets/Autistic+Spectrum+Disorder+Screening+Data+for+Adolescent+++" rel=" noopener" target="_blank" src="https://archive.ics.uci.edu/ml/datasets/Autistic+Spectrum+Disorder+Screening+Data+for+Adolescent+++" name="cta=External Link Click">Autistic Spectrum Disorder Screening Data for Adolescent Data Set</a>’ from the University of California, Irvine.</p>

```python
data = arff.loadarff('./Autism-Adolescent-Data.arff')
df = pd.DataFrame(data[0])
list(df)
df.head()
```

```python
str_df = df.select_dtypes([np.object])
str_df = str_df.stack().str.decode('utf-8').unstack()
for col in str_df:
    df[col] = str_df[col]
```

```python
df.head()
```

<p class="selectionShareable">This is a relatively small dataset, so random forest is the perfect model because it uses bagging, which I explained earlier in this tutorial. Before we build the model, we need to make some changes to the data in order to make it ready for the model.</p>

<p class="selectionShareable">Let’s begin with lowercasing and one-hot encoding the categorical variables so that we can turn the categorical variables to numeric. Let’s make the following changes:</p>
<ul> 
  <li><p class="selectionShareable"><span class="hs_cos_wrapper hs_cos_wrapper_meta_field hs_cos_wrapper_type_rich_text" data-hs-cos-general-type="meta_field" data-hs-cos-type="rich_text" id="hs_cos_wrapper_post_body" style="">Lowercase all the text</span></p></li> 
  <li><p class="selectionShareable"><span class="hs_cos_wrapper hs_cos_wrapper_meta_field hs_cos_wrapper_type_rich_text" data-hs-cos-general-type="meta_field" data-hs-cos-type="rich_text" id="hs_cos_wrapper_post_body" style="">All ‘yes’ = 1</span></p></li> 
  <li><p class="selectionShareable"><span class="hs_cos_wrapper hs_cos_wrapper_meta_field hs_cos_wrapper_type_rich_text" data-hs-cos-general-type="meta_field" data-hs-cos-type="rich_text" id="hs_cos_wrapper_post_body" style="">All ‘no’ = 0</span></p></li> 
  <li><p class="selectionShareable"><span class="hs_cos_wrapper hs_cos_wrapper_meta_field hs_cos_wrapper_type_rich_text" data-hs-cos-general-type="meta_field" data-hs-cos-type="rich_text" id="hs_cos_wrapper_post_body" style="">Female = 1</span></p></li> 
  <li><p class="selectionShareable"><span class="hs_cos_wrapper hs_cos_wrapper_meta_field hs_cos_wrapper_type_rich_text" data-hs-cos-general-type="meta_field" data-hs-cos-type="rich_text" id="hs_cos_wrapper_post_body" style="">Male = 0</span></p></li> 
 </ul>

```python

df = df.apply(lambda x: x.astype(str).str.lower())
df = df.replace('yes', 1)
df = df.replace('no', 0)
df = df.replace('f', 1)
df = df.replace('m', 0)
```

```python
df
```

<h2 class="selectionShareable">Step 3: Subset the data</h2>
<p class="selectionShareable">Our new dataset should only have the variables that we will be using to build the model.</p>

```python

xVar = list(df.loc[:,'A1_Score':'A10_Score']) + ['gender'] + ['jundice'] + ['austim']
yVar = df.iloc[:,20]
df2 = df[xVar]
```

<h2 class="selectionShareable">Step 4: Split the data into train and test sets</h2>
<p class="selectionShareable">We will build the model on the training set and check the accuracy of the model by using it on the testing set.</p>

```python

X_train, X_test, y_train, y_test = train_test_split(df2, yVar, test_size=0.2)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

<p class="selectionShareable">In the code above, the data is split in a way that 80% of the variables fall under the training set and 20% of the variables are used for <a href="https://www.datascience.com/resources/white-papers/predictive-model-testing" rel=" noopener" target="_blank" src="https://www.datascience.com/resources/white-papers/predictive-model-testing" name="cta=External Link Click">testing the model</a>. Our resulting training set has 83 observations and the testing set has 21 observations.</p>

<h2 class="selectionShareable">Step 5: Build a Random Forest Classifier</h2>
    <p class="selectionShareable">Next, let’s build the random forest classifier and fit the model to the data. We will use the default parameters for the model.</p>

To identify anomalies, we can quickly compute summary statistics.

```python

clf = RandomForestClassifier(n_jobs=2, random_state=0)

clf.fit(X_train, y_train)
RandomForestClassifier(
    bootstrap=True, 
    class_weight=None, 
    criterion='gini',
    max_depth=None, 
    max_features='auto', 
    max_leaf_nodes=None,
    min_impurity_split=1e-07, 
    min_samples_leaf=1,
    min_samples_split=2, 
    min_weight_fraction_leaf=0.0,
    n_estimators=10, 
    n_jobs=2, 
    oob_score=False, 
    random_state=0,
    verbose=0, 
    warm_start=False)
```
